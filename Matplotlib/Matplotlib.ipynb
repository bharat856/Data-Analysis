```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 10, 100)  # Ensures x > 0 for log plot

fig, axs = plt.subplots(2, 2, figsize=(10, 8))  # Correct function and figure size

axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("Sine Wave")

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title("Cosine Wave")

axs[1, 0].plot(x, np.random.rand(100))  # Corrected random plot
axs[1, 0].set_title("Random")

axs[1, 1].plot(x, np.log(x))  # Log function requires x > 0
axs[1, 1].set_title("Log")

plt.tight_layout()  # Adjusts layout for better visibility
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the grid
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

# Create figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot surface
ax.plot_surface(X, Y, Z, cmap='viridis')

plt.show()
```

```python
import random
import matplotlib.pyplot as plt

# Initialize counts
h_t = [0, 0]

# Create figure
plt.ion()  # Turn on interactive mode
fig, ax = plt.subplots()

for _ in range(1000):
    # Flip a coin
    h_t[random.randint(0,1)] += 1

    # Clear previous bars before replotting
    ax.clear()
    ax.bar(['Heads', 'Tails'], h_t, color=['red', 'blue'])
    ax.set_ylim(0, max(h_t) + 50)  # Adjust y-axis dynamically

    plt.pause(0.001)  # Pause for smooth animation

plt.ioff()  # Turn off interactive mode
plt.show()  # Ensure the final plot stays visible
    
```
