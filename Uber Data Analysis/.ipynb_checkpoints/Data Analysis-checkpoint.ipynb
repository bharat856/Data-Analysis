```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load dataset
file_path = "uber_data.csv"  # Ensure correct file path

df = pd.read_csv(file_path)

# Convert datetime columns
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['hour'] = df['tpep_pickup_datetime'].dt.hour

# Use dark background style
plt.style.use('dark_background')

# 1. Trip Distance vs. Fare Amount (Scatter Plot)
plt.figure(figsize=(10, 6))
plt.scatter(df['trip_distance'], df['fare_amount'], alpha=0.5)
plt.xlabel("Trip Distance (miles)")
plt.ylabel("Fare Amount ($)")
plt.title("Trip Distance vs. Fare Amount")
plt.show()

# 2. Distribution of Trip Distances (Histogram)
plt.figure(figsize=(10, 6))
plt.hist(df['trip_distance'], bins=50, edgecolor='white')
plt.xlabel("Trip Distance (miles)")
plt.ylabel("Frequency")
plt.title("Distribution of Trip Distances")
plt.xlim(0, np.percentile(df['trip_distance'], 99))  # Avoid extreme outliers
plt.show()

# 3. Total Fare Distribution (Histogram)
plt.figure(figsize=(10, 6))
plt.hist(df['total_amount'], bins=50, edgecolor='white')
plt.xlabel("Total Fare Amount ($)")
plt.ylabel("Frequency")
plt.title("Total Fare Distribution")
plt.xlim(0, np.percentile(df['total_amount'], 99))  # Avoid extreme outliers
plt.show()

# 4. Number of Rides per Hour (Bar Chart)
plt.figure(figsize=(10, 6))
df['hour'].value_counts().sort_index().plot(kind='bar', color='cyan')
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Rides")
plt.title("Number of Rides per Hour")
plt.show()

# 5. Heatmap of Pickup Locations (Hexbin Plot)
plt.figure(figsize=(10, 6))
plt.hexbin(df['pickup_longitude'], df['pickup_latitude'], gridsize=100, cmap='coolwarm', mincnt=1)
plt.xlabel("Pickup Longitude")
plt.ylabel("Pickup Latitude")
plt.title("Heatmap of Pickup Locations")
plt.colorbar(label='Number of Rides')
plt.show()

# 6. Payment Type Distribution (Pie Chart)
payment_counts = df['payment_type'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=140, colors=['red', 'blue', 'green', 'yellow'])
plt.title("Payment Type Distribution")
plt.show()

# 7. Number of Passengers per Trip (Bar Chart)
plt.figure(figsize=(10, 6))
df['passenger_count'].value_counts().sort_index().plot(kind='bar', color='orange')
plt.xlabel("Number of Passengers")
plt.ylabel("Count of Trips")
plt.title("Number of Passengers per Trip")
plt.show()

# 8. Tip Amount vs. Total Fare (Scatter Plot)
plt.figure(figsize=(10, 6))
plt.scatter(df['total_amount'], df['tip_amount'], alpha=0.5, color='magenta')
plt.xlabel("Total Fare Amount ($)")
plt.ylabel("Tip Amount ($)")
plt.title("Tip Amount vs. Total Fare")
plt.show()

# 9. Total Revenue Per Hour (Line Plot)
df.groupby('hour')['total_amount'].sum().plot(kind='line', marker='o', figsize=(10, 6), color='lime')
plt.xlabel("Hour of the Day")
plt.ylabel("Total Revenue ($)")
plt.title("Total Revenue Per Hour")
plt.grid()
plt.show()
```
